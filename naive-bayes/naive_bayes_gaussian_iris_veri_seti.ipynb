## Veri setinin alınması

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

## Bağımlı ve Bağımsız Değişkenleri Birbirinden Ayırma



```python
X_iris = iris.drop('species', axis = 1)
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

## Naive Bayes Gaussian Kullanarak Modelin Oluşturulması

```python
from sklearn.model_selection import train_test_split

X_egitim, X_test, y_egitim, y_test = train_test_split(X_iris, y_iris, random_state = 1, test_size = 0.2) # iris veri setini eğitim ve test verisi olarak 2'ye ayırıyoruz.
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()

model.fit(X_egitim, y_egitim) # modelimizi fit ederek oluşumunu tamamlıyoruz.
```

```python
from sklearn.metrics import accuracy_score

y_model = model.predict(X_test) # oluşturulan modele test verilerimizi veriyoruz.

accuracy_score(y_test, y_model) #modelin tahminlerini olması gereken ifadeler ile karşılaştırarak modelin doğruluk oranına ulaşıyoruz.
```
