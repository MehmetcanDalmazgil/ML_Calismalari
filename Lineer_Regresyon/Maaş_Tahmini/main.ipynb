**Kütüphanelerin Eklenmesi**

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from word2number import w2n
```

**Verinin Excel Dosyasından Okunması**

```python
# Verimizi excel dosyası içerisinden alıyoruz. 
veri = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Lineer Regresyon/Maaş Tahmini/maas.xlsx")
veri
```

**Veri Ön İşleme**


```python
# Deneyim sütunu içerisinde yer alan boşlukları dolduruyoruz.
veri.deneyim = veri.deneyim.fillna("zero")
veri
```

```python
# Deneyim sütunu içerisinde yer alan değerleri sayıya çeviriyoruz.
veri.deneyim = veri.deneyim.apply(w2n.word_to_num)
veri
```

```python
# Teknik mülakat puanı içerisinde yer alan boşlukları doldurmak için puan medyanını hesaplıyoruz.
import math
medyan_teknik = math.floor(veri['teknikmulakat'].mean())
medyan_teknik
```

```python
# Teknik mülakat sütunu içerisinde yer alan boşlukları dolduruyoruz.
veri['teknikmulakat'] = veri['teknikmulakat'].fillna(medyan_teknik)
veri
```

```python
# Verimiz kullanıma hazır :)
```

**Lineer Regresyon Modelinin Oluşturulması**

```python
# Modelimizi elimizdeki verileri kullanarak eğitiyoruz.
model = linear_model.LinearRegression()
model.fit(veri[['deneyim','teknikmulakat','sozelmulakat']],veri['maas'])
```

```python
# Modelimize tabloda bulunan bir veri kontrol amaçlı verildiğinde modelin yakın bir tahmin gerçekleştirdiği görülmektedir.
model.predict([[0,8,9]])
```

```python
# Maaş tahmini örneği
model.predict([[2,9,6]])
```

```python
# m1.x1 + m2.x2 + m3.x3 + b = y
# Lineer regresyon modelinin ağırlık değerleri
model.coef_
```

```python
# Lineer regresyon modelinin sabit değeri
model.intercept_
```
