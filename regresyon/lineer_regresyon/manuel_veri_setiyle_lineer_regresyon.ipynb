## Manuel Oluşturulan Veri Setiyle Lineer Regresyon

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(22) # program her çalıştığında aynı random sayıların oluşturulmasını sağlıyoruz.

x = 10 * rng.rand(50) # 0-1 arasında sayılar oluştuğundan 10 ile çarpıyoruz. İşlem sonucunda 50 elemanlık bir numpy dizi oluşmaktadır.
x
```

```python
y = 2 * x - 1 + rng.randn(50) # x'e bağlı bir bilinmeyenli denklem oluşturuyoruz.
plt.scatter(x,y)
# x değişkenine bağlı olarak y değişkeninde meydana gelen değişime ait grafik aşağıdadır.
```

```python
# problem değişkenleri sayısal ifadeler oolduğu için regresyon kullanılacaktır.
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# X değerlerini modele verebilmek için en az iki boyutlu bir matrise ihtiyaç vardır.
X = x[:, np.newaxis] # 2 boyutlu öznitelik matrisi
X.shape
```

```python
model.fit(X,y) # modelimizi çalıştırıyoruz.

print(model.coef_) # kaysayı değeri
print(model.intercept_) # sabit kaysayı değeri
```

```python
X_fit = np.linspace(-1, 11) # değerler arasını 50 parçaya bölmektedir.
X_fit = X_fit[:, np.newaxis] # rastgele test verisi oluşturarak veriyi modele verebilmemiz adına iki boyutlu hale getiriyoruz.
X_fit
```

```python
y_fit = model.predict(X_fit) # modelimize hazırladığımız test verilerini veriyoruz.
plt.scatter(X,y)
plt.plot(X_fit, y_fit)
# bu şekilde veri setine göre regresyon doğrusu çizilmiş olmaktadır.
```
